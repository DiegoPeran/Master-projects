# Exercise 1

We have created the following function to return the words of a string. First, the function removes all characters of the original string that are not letters or ' (to include words like don't, it's, etc.). Second, the function splits the string (clean out of forbidden characters) using the space separator (i.e. it separates words). Third, if a particular string (word), was contained in the original string, the function includes it in a new list. This last point is included to avoid inserting words in certain situations. For instance we find the following string on a message: "cantaloupe.srv.cs.cmu.edu"; our function would delete the forbidden characters and it would become: "cantaloupesrvcscmuedu"; if we do not incldue that last if sentence, this word would be included in the list without sense. It has also inconvenients; for instance, a word like "non-umpleasant" will be excluded, since the function will remove "-" and the word "nonumpleasant" would not be included in the original string. 

```python
def returnWords(m):
    m=m.lower()
    abc="ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz '"
    newm=""
    words=list()
    for j in range(len(m)):
        if m[j] in abc:
            newm+=m[j]
    for i in range(len(newm.split())):
        if newm.split()[i] in m:
            words.append(newm.split()[i])
    
    return words
```

```python
message="hola! Don Pepito, hola << Don Jose&"
message2 = "In a hole in the ground, there lived a Hobbit. He said he'd make some chocolate. He was non-umpleasant"
print("First message:",returnWords(message),"Second message:",returnWords(message2),sep="\n")
```

# Exercise 2

We are creating the function countWords, that returns a default dictionary with lambda 0. This dictionary includes the number of times a word is repeated in a sentence (or a message). In addition, we are testing it with the previous message. We use a default dictionary with lambda:0 since it will be convenient in following parts of this task.

```python
from collections import defaultdict

def countWords(s):
    # s must be a string (a sentence)
    sprov = s.lower()
    sprov = returnWords(sprov)
    dd = defaultdict(lambda: 0)
    for words in sprov:
        dd[words] += 1
    return(dd)
```

We observe "directories" is a list of strings (i.e. elements indicating the name of each subdirectory). Therefore, we can use the function random.choice() of the module "random" to pick randomly an element of this list. We have created a copy of the "directories" list, in order to not modify it. Then, we use a for loop to pick randomly an element from the "directories2" list and we remove it from the list (in order to not select the same one more than once). We iterate 5 times, to get 5 names of directories. Finally, we check the selected ones.

```python
# print('The subdirectories are:', directories) # Checking the names of the elements

import random
import copy

directories2 = copy.deepcopy(directories)
selected_sdir = list()

for i in range(5):
    element = random.choice(directories2)
    selected_sdir.append(element)
    indextodelete = directories2.index(element)
    del(directories2[indextodelete])

# print('The subdirectories are:', directories) # It has not changed
# print('The subdirectories are:', directories2) # It has changed and does not contain the selected directories
print(selected_sdir)
```

```python
messages=list()
for element in selected_sdir:
    for k in range(8):
        messages.append(getMessage(prefix + element + "\\" + os.listdir(prefix+element)[k]))

print(messages[0]) # Check the first element out of the 40 messages selected
# print(messages[40]) # Check that we get an error and there are not more than 40 messages
```

# Exercise 5

For creating a matrix with the word counts, we proceed as follows. First, we create a matrix of 0's with the number of rows equal to the number of messages (in this case, 40) and as columns as different words are in our 40 messages (it varies since it is generated randomly). Then we iterate first for every message, and then for every column (different word). Now we access the i-th dictionary of our list of dictionaries "countMessages", and we introduce in the j-th column, the number of counts of that word in that message. 

We have done this process with two different notations, to work a bit further. The first one consists on iterate over the words contained in the set "uniqueWords" and thus it is necessary to label "j" with increasing numbers to introduce it in the matrix. Then we acces the dictionary i-th and ask for the value of the key="word" (if it is not included in that dictionary, since it is a default dictionary with lambda:0, it leaves the 0 in the matrix). The second one iterates from 0 to the (number of total unique words-1) for every message. Then, we have to acces the dictionary with the corresponding word of the set "uniqueMessages", for which we have to transform it to a list, and access the element j-th.

```python
import numpy as np

# Procedure 1: iterating over the "words"

M1 = np.zeros((40,len(uniqueMessages)),dtype=int)
for i in range(40):
    for word in uniqueMessages:
        j = list(uniqueMessages).index(word)
        M1[i,j] = countMessages[i][word]
```

```python
# Procedure 2: iterating over the numbers
M = np.zeros((40,len(uniqueMessages)),dtype=int)
for i in range(40):
    for j in range(len(uniqueMessages)):
            M[i,j] = countMessages[i][list(uniqueMessages)[j]]
```

In the following line, we are checking both procedures give us the same results. First we check the sum of the columns is equal for both matrices (we check we have a number of Trues equal to the number of different words). Second, we check the sum of the rows is equal for both matrices.

```python
print(sum(M1.sum(axis=0)==M.sum(axis=0)*1),"\n") # It has to be equal to the number of different words (number of columns)
print(M1.sum(axis=1)==M.sum(axis=1)) # Check we have 40 Trues, all the row sums are equal for both matrices
```

# Exercise 5 (second part)

In order to compute the matrix of ratios, we can use broadcasting properties of Python. We can compute the division of all elements of a row by the total sum of that row dividing the matrix (40x(number of unique words)) by the vector of total sums (40x1). This operation will act as it were (40x(number of unique words)) / "(40x(number of unique words))" where the second denominator term is a matrix containing repetitions of the vector of total sums in each colum. Therefore, it will do the division element by element, which results in a matrix where all elements of the same row are divided by the total sum of the elements of that row (i.e. by the total number of words in that message --> proportions of each word in each message).

We have to introduce the argument "keepdims" to keep dimension (4,1). This is not necessary when we use row vectors, since (1,4) is equivalent to (4,) in these terms.

```python
Mratios=M/M.sum(axis = 1, keepdims=True) 
print(Mratios)
```
